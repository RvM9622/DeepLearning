```python
import pandas as pd

df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(['id', 'date', 'zipcode','yr_built', 'yr_renovated', 'lat', 'long'], axis = 1)
```

```python
df.info()
```

```python
x = df.drop('price', axis = 1)
y = df['price']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 0)
```

#Build ANN Model

```python
from tensorflow import keras

model = keras.models.Sequential([keras.layers.Dense(100, activation= 'relu', input_shape=(13,)),
                                 keras.layers.Dense(100, activation = 'relu'),
                                 keras.layers.Dense(100, activation = 'relu'),
                                 keras.layers.Dense(1, activation = 'linear')])
```

```python
model.compile(optimizer='adam', loss = 'mean_squared_error', metrics=['R2Score'])
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
model.evaluate(x_test, y_test)
```
